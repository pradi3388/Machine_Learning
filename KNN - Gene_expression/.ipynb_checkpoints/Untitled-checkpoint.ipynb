### Train|Test Split and Scaling Data

```python
X = df.drop('Cancer Present', axis = 1)
y = df['Cancer Present']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
```

```python
scaled_X_train = scalar.fit_transform(X_train)
scaled_X_test = scalar.transform(X_test)
```

### Understanding KNN and Choosing K Value (Taking K=1)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=1)
```

```python
knn_model.fit(scaled_X_train, y_train)
```

```python
y_pred = knn_model.predict(scaled_X_test)
```

### Model Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix, classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
plot_confusion_matrix(estimator=knn_model, X=scaled_X_test, y_true=y_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
error =1- accuracy_score(y_test,y_pred)
error
```

### Elbow Method for Choosing Reasonable K Values

**NOTE: This uses the test set for the hyperparameter selection of K.**

```python
# we have to select k value where error rate is minimum

error_rate = []

for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(scaled_X_train, y_train)
    
    y_pred = knn_model.predict(scaled_X_test)
    
    error= 1-accuracy_score(y_test, y_pred)
    error_rate.append(error)
```

```python
error_rate
```

```python
plt.plot(range(1,30),error_rate)
plt.xlabel('K-Value')
plt.ylabel('Error Rate')
```

From graph we can see after K=6 error rate is not much improving, So we can chose 6, but still leads error if we wrongly selected manually. So we can use cross validation instead of Elbow method

### Final Model

We just saw that our  GridSearch recommends a K=14 (in line with our alternative Elbow Method). Let's now use the PipeLine again, but this time, no need to do a grid search, instead we will evaluate on our hold-out Test Set.

```python
scalar = StandardScaler()
knn_model21= KNeighborsClassifier(n_neighbors=21)
operation = [('scalar', scalar), ('knn_model21', knn_model21)]
```

```python
pipe = Pipeline(operation)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
